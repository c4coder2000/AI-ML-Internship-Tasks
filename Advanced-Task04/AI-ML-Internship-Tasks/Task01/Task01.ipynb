Now visualizing the dataset

```python
import matplotlib.pyplot as plt
sns.pairplot(df, hue='species', palette='viridis', height=2.5)
plt.suptitle("Pairwise Relationships in Iris Dataset", y=1.02)
plt.show()
```

Plotting histogram

```python
plt.figure(figsize=(12, 8))


features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data=df,
                 x=feature,
                 hue='species',
                 palette='viridis',
                 element='step',
                 alpha=0.6,
                 bins=15,
                 kde=True)
    plt.title(f'Distribution of {feature.replace("_", " ").title()}')
    plt.xlabel(feature.replace("_", " ").title() + ' (cm)')

plt.suptitle("Histograms of Iris Features by Species", y=1.02)
plt.tight_layout()
plt.show()
```

Box Plot for detecting outliers

```python
plt.figure(figsize=(10, 6))
df.boxplot()
plt.title('Box Plots of All Iris Features')
plt.ylabel('Centimeters (cm)')
plt.xticks(rotation=45)
plt.show()
```

The points outside the whiskers are outliers
